```python
def get_both_plots(data_name):
    kaggle_name = f"kaggle_{data_name}"
    base_path = pathlib.Path("openfe_experiments")
    path1 = base_path / f"openfe_{kaggle_name}_10k_mcf_2_db_fb_False_grid_search"
    path2 = base_path / f"openfe_{kaggle_name}_10k_mcf_2_db_fb_False_grid_search_blend"
    plot1 = path1 / kaggle_name / "openfe_scores_plot.png"
    plot2 = path2 / kaggle_name / "openfe_scores_plot.png"
    return str(plot1), str(plot2)
```

```python
def compare_plots(data_name, single_lgbm_plot, lgbm_blend_plot, figsize=(15, 9)):
    """
    Display two plots side by side for comparison.
    
    Args:
        data_name: Name of the dataset
        figsize: Figure size as (width, height)
    """
    
    # Create figure with subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=figsize)
    
    # Load and display the images
    img1 = mpimg.imread(single_lgbm_plot)
    img2 = mpimg.imread(lgbm_blend_plot)
    
    ax1.imshow(img1)
    ax1.set_title(f'{data_name} - Single LGBM', fontsize=12)
    ax1.axis('off')
    
    ax2.imshow(img2)
    ax2.set_title(f'{data_name} - LGBM Blend', fontsize=12)
    ax2.axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python
for data_name in datasets:
    try:
        print(f"\n\nComparing plots for dataset: {data_name}")
        single_lgbm_plot, lgbm_blend_plot = get_both_plots(data_name)
        compare_plots(data_name, single_lgbm_plot, lgbm_blend_plot)
    except Exception as e:
        print(f"An error occurred while processing {data_name}: {e}")
```
